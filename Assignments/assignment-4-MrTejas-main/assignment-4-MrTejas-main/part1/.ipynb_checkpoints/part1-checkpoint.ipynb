```python
import numpy as np
import time
import matplotlib.pyplot as plt
import cv2
```

```python
epsilon = 1+1e-12
```

```python
def minmax_scale(ar):
    br = (ar-np.min(ar))*255/(np.max(ar)-np.min(ar))
    br = br.astype(np.uint8)
    return br
```

```python
# def dft1(x):
#     n = x.shape[0]
#     i = np.arange(n)
#     j = i.reshape((n,1))
#     f = np.exp(-2j*np.pi*i*j/n)
#     X = np.dot(x,f)
#     return X
# def dft2(x):
#     m,n = x.shape
#     X = np.zeros_like(x, dtype=np.complex128)
#     for i in range(m):
#         X[i, :] = dft1(x[i, :])
#     for j in range(n):
#         X[:, j] = dft1(X[:, j])
#     return X


def dft1(x):
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    X = np.dot(e, x)
    return X

def dft2(x):
    M, N = x.shape
    X = np.zeros_like(x, dtype=np.complex128)

    # Apply 1D DFT along each row
    for i in range(M):
        X[i, :] = dft1(x[i, :])

    # Apply 1D DFT along each column
    for j in range(N):
        X[:, j] = dft1(X[:, j])

    return X

def fft1(x):
    N = len(x)
    if N <= 1:
        return x
    else:
        even = fft1(x[0::2])
        odd = fft1(x[1::2])
        factor = np.exp(-2j * np.pi * np.arange(N) / N)
        return np.concatenate([even + factor[:N // 2] * odd, even + factor[N // 2:] * odd])

def fft2(x):
    M, N = x.shape
    padded_M = 2 ** int(np.ceil(np.log2(M)))
    padded_N = 2 ** int(np.ceil(np.log2(N)))

    # Pad the input array with zeros
    padded_x = np.zeros((padded_M, padded_N), dtype=x.dtype)
    padded_x[:M, :N] = x

    # Compute FFT along rows
    fft_rows = np.zeros_like(padded_x, dtype=np.complex128)
    for i in range(padded_M):
        fft_rows[i, :] = fft1(padded_x[i, :])

    # Compute FFT along columns
    fft_cols = np.zeros_like(padded_x, dtype=np.complex128)
    for j in range(padded_N):
        fft_cols[:, j] = fft1(fft_rows[:, j])

    return fft_cols

```

```python
ar_size = np.arange(7)
ar_size = 2**(ar_size+7)
low = 0
high = 100

run_time_dft = []
run_time_fft = []

for n in ar_size:
    ar = np.random.randint(low=low, high=high, size=n)
    t1 = time.time()
    ar_dft = dft1(ar)
    t2 = time.time()
    run_time_dft.append(t2-t1)
    t1 = time.time()
    ar_fft = fft1(ar)
    t2 = time.time()
    run_time_fft.append(t2-t1)

plt.plot(ar_size, np.array(run_time_dft), label='DFT')
plt.plot(ar_size, np.array(run_time_fft), label='FFT')
plt.xlabel('Array size (n)')
plt.ylabel('Run-time (sec)')
plt.title('Run-time vs array size for DFT and FFT')
plt.legend()
plt.grid()
plt.show()

```

```python
# visualizes the DFT with log spectrum
def visualize_dft_pattern(dft_result, axs):
    magnitude_spectrum = np.log1p(np.abs(dft_result))
    axs.imshow(magnitude_spectrum, cmap='gray')
```

```python
def idealLPF(image, D0):
    # Convert the image to grayscale
    gray_image = color.rgb2gray(image)

    # Compute the 2D FFT of the image
    image_fft = fft2(gray_image)

    # Create an ideal low-pass filter in the frequency domain
    M, N = image.shape
    u = np.fft.fftfreq(M)
    v = np.fft.fftfreq(N)
    U, V = np.meshgrid(u, v)
    D = np.sqrt(U**2 + V**2)

    # Apply the ideal low-pass filter
    H = np.zeros_like(D)
    H[D <= D0] = 1

    # Apply the filter in the frequency domain
    filtered_image_fft = image_fft * H

    # Inverse FFT to get the filtered image in spatial domain
    filtered_image = np.abs(np.fft.ifft2(filtered_image_fft))

    return filtered_image
```

```python
image = plt.imread('./pic.png')

# Set the cutoff frequency D0 (you can adjust this value)
D0 = 30

# Apply the ideal low-pass filter
filtered_image = idealLPF(image, D0)

# Display the original and filtered images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title(f'Ideal LPF (D0={D0}) Filtered Image')

plt.show()
```
