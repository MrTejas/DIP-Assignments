```python
from matplotlib import pyplot as plt
import cv2
```

```python
# function to display an image
def display_image(img,s):
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(s)
```

```python
# function to save an image
def save_image(img,s):
    name = './../images/'+s+'.jpg'
    cv2.imwrite(name,img)
```

```python
# function to display images in 4x4 subplot
def display4Image_cv2(im1, im2, im3, im4):
    im1 = cv2.cvtColor(im1,cv2.COLOR_BGR2RGB)
    im2 = cv2.cvtColor(im2,cv2.COLOR_BGR2RGB)
    im3 = cv2.cvtColor(im3,cv2.COLOR_BGR2RGB)
    im4 = cv2.cvtColor(im4,cv2.COLOR_BGR2RGB)
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
    fig.suptitle('4 cv2_images Plotted using Subplot')
    ax1.imshow(im1)
    ax2.imshow(im2)
    ax3.imshow(im3)
    ax4.imshow(im4)
```

```python
# function to convert an image into greyscale using opencv
def col_to_greyscale_cv2(img):
    dim = img.shape
    n,m,c = dim
    for i in range(n-1):
        for j in range(m-1):
            r,g,b = img[i,j]
            avg = int((int(r)+int(b)+int(g))/3.0)
            img[i,j] = (avg,avg,avg)
    return img
```

```python
# DISPLAYING IMAGES IN 4X4 SUBPLOT

loc1 = './../images/im1.jpg'
loc2 = './../images/im2.jpg'
loc3 = './../images/im3.jpg'
loc4 = './../images/im4.jpg'

img1 = cv2.imread(loc1)
img2 = cv2.imread(loc2)
img3 = cv2.imread(loc3)
img4 = cv2.imread(loc4)

display4Image_cv2(img1,img2,img3,img4)
```

```python
# CONVERTING IMAGES INTO GREYSCALE USING CV2 AND SAVING THEM

# image addresses in the directory
loc1 = './../images/im1.jpg'
loc2 = './../images/im2.jpg'
loc3 = './../images/im3.jpg'
loc4 = './../images/im4.jpg'

# loading images using opencv
img1 = cv2.imread(loc1)
img2 = cv2.imread(loc2)
img3 = cv2.imread(loc3)
img4 = cv2.imread(loc4)

# converting images to greyscale
img1 = col_to_greyscale_cv2(img1)
img2 = col_to_greyscale_cv2(img2)
img3 = col_to_greyscale_cv2(img3)
img4 = col_to_greyscale_cv2(img4)

# displaying all the images
display4Image_cv2(img1,img2,img3,img4)

# saving images in the ./../images/ folder
save_image(img1,'im1_greyscale_cv')
save_image(img2,'im2_greyscale_cv')
save_image(img3,'im3_greyscale_cv')
save_image(img4,'im4_greyscale_cv')
```
