```python
import skimage
from skimage import io
from matplotlib import pyplot as plt
```

```python
# function to display an image
def display_image(img,s):
    io.imshow('./../images/im1.jpg')
    io.show()
    print(s)
```

```python
# function to save an image
def save_image(img,s):
    name = './../images/'+s+'.jpg'
    io.imsave(name,img)
```

```python
# function to display images in 4x4 subplot
def display4Image_Scikit(im1, im2, im3, im4):
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
    fig.suptitle('4 Scikit_images Plotted using Subplot')
    ax1.imshow(im1)
    ax2.imshow(im2)
    ax3.imshow(im3)
    ax4.imshow(im4)
```

```python
# function to convert an image into greyscale using Scikit
def col_to_greyscale_Scikit(img):
    (n,m,c) = img.shape
    
    for i in range(n-1):
        for j in range(m-1):
            r,g,b = img[i,j]
            avg = int((int(r)+int(b)+int(g))/3.0)
            img[i,j] = (avg,avg,avg)
    return img
```

```python
# DISPLAYING IMAGES IN 4X4 SUBPLOT

loc1 = './../images/im1.jpg'
loc2 = './../images/im2.jpg'
loc3 = './../images/im3.jpg'
loc4 = './../images/im4.jpg'

img1 = io.imread(loc1)
img2 = io.imread(loc2)
img3 = io.imread(loc3)
img4 = io.imread(loc4)

display4Image_Scikit(img1,img2,img3,img4)
```

```python
# CONVERTING IMAGES INTO GREYSCALE USING Scikit AND SAVING THEM

# image addresses in the directory
loc1 = './../images/im1.jpg'
loc2 = './../images/im2.jpg'
loc3 = './../images/im3.jpg'
loc4 = './../images/im4.jpg'

# loading images using Scikit
img1 = io.imread(loc1)
img2 = io.imread(loc2)
img3 = io.imread(loc3)
img4 = io.imread(loc4)

# converting images to greyscale
img1 = col_to_greyscale_Scikit(img1)
img2 = col_to_greyscale_Scikit(img2)
img3 = col_to_greyscale_Scikit(img3)
img4 = col_to_greyscale_Scikit(img4)

# displaying all the images
display4Image_Scikit(img1,img2,img3,img4)

# saving images in the ./../images/ folder
save_image(img1,'im1_greyscale_Scikit')
save_image(img2,'im2_greyscale_Scikit')
save_image(img3,'im3_greyscale_Scikit')
save_image(img4,'im4_greyscale_Scikit')
```
