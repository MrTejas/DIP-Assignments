```python
from matplotlib import pyplot as plt
from PIL import Image
```

```python
# function to display an image
def display_image(img,s):
    display(img)
    print(s)
```

```python
# function to save an image
def save_image(img,s):
    name = './../images/'+s+'.jpg'
    img.save(name)
```

```python
# function to display images in 4x4 subplot
def display4Image_PIL(im1, im2, im3, im4):
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
    fig.suptitle('4 PIL_images Plotted using Subplot')
    ax1.imshow(im1)
    ax2.imshow(im2)
    ax3.imshow(im3)
    ax4.imshow(im4)
```

```python
# function to convert an image into greyscale using PIL
def col_to_greyscale_PIL(img):
    [n,m] = img.size
    px = img.load()
    
    for i in range(n-1):
        for j in range(m-1):
            r,g,b = px[i,j]
            avg = int((int(r)+int(b)+int(g))/3.0)
            px[i,j] = (avg,avg,avg)
    return img
```

```python
# DISPLAYING IMAGES IN 4X4 SUBPLOT

loc1 = './../images/im1.jpg'
loc2 = './../images/im2.jpg'
loc3 = './../images/im3.jpg'
loc4 = './../images/im4.jpg'

img1 = Image.open(loc1)
img2 = Image.open(loc2)
img3 = Image.open(loc3)
img4 = Image.open(loc4)

display4Image_PIL(img1,img2,img3,img4)
```

```python
# CONVERTING IMAGES INTO GREYSCALE USING PIL AND SAVING THEM

# image addresses in the directory
loc1 = './../images/im1.jpg'
loc2 = './../images/im2.jpg'
loc3 = './../images/im3.jpg'
loc4 = './../images/im4.jpg'

# loading images using PIL
img1 = Image.open(loc1)
img2 = Image.open(loc2)
img3 = Image.open(loc3)
img4 = Image.open(loc4)

# converting images to greyscale
img1 = col_to_greyscale_PIL(img1)
img2 = col_to_greyscale_PIL(img2)
img3 = col_to_greyscale_PIL(img3)
img4 = col_to_greyscale_PIL(img4)

# displaying all the images
display4Image_PIL(img1,img2,img3,img4)

# saving images in the ./../images/ folder
save_image(img1,'im1_greyscale_PIL')
save_image(img2,'im2_greyscale_PIL')
save_image(img3,'im3_greyscale_PIL')
save_image(img4,'im4_greyscale_PIL')
```
